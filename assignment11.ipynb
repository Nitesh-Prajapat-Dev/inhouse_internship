```python
from statistics import mode

arr1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

arr2 = np.array([
    [7, 8, 9],
    [10, 11, 12]
])

# Average Array
avg = (arr1 + arr2) / 2

print("Average Array:\n")
print(avg)

# Combine Arrays
combined = np.concatenate((arr1, arr2))

# Mean
print("\nMean =", np.mean(combined))

# Median
print("Median =", np.median(combined))

# Mode
print("Mode =", mode(combined.flatten()))
```

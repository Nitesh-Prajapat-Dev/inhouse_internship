```python
import pandas as pd

data = {
    "Name": ["Nitesh", "Rahul"],
    "Marks": [85, 78]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    "Name": ["Aman"],
    "Marks": [90]
})

df = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print(df)
```

```python
import csv

with open("address_book.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow(["Name", "Address", "Mobile", "Email"])

    for i in range(3):

        print(f"\nEnter Details for Person {i+1}")

        name = input("Enter Name: ")
        address = input("Enter Address: ")
        mobile = input("Enter Mobile Number: ")
        email = input("Enter Email: ")

        writer.writerow([name, address, mobile, email])

print("\nCSV File Created Successfully")
```

```python
import csv

with open("address_book.csv", "r") as file:

    reader = csv.reader(file)

    print("\nAddress Book Data:\n")

    for row in reader:
        print(row)
```

```python
import csv

with open("address_book.csv", "a", newline="") as file:

    writer = csv.writer(file)

    print("\nEnter New Person Details")

    name = input("Enter Name: ")
    address = input("Enter Address: ")
    mobile = input("Enter Mobile Number: ")
    email = input("Enter Email: ")

    writer.writerow([name, address, mobile, email])

print("\nData Appended Successfully")
```

```python
import sqlite3

conn = sqlite3.connect("college.db")

cur = conn.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS student (
    id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    marks INTEGER
)
""")

print("Table Created Successfully")
```

```python
cur.execute("INSERT INTO student VALUES (1, 'Nitesh', 'BCA', 85)")
cur.execute("INSERT INTO student VALUES (2, 'Rahul', 'BCA', 78)")
cur.execute("INSERT INTO student VALUES (3, 'Aman', 'BCA', 90)")

conn.commit()

print("Records Inserted")
```

```python
cur.execute("SELECT * FROM student")

data = cur.fetchall()

print("\nStudent Records:\n")

for row in data:
    print(row)
```

```python
cur.execute("UPDATE student SET marks = 95 WHERE id = 1")

conn.commit()

print("\nData Updated")
```

```python
cur.execute("DELETE FROM student WHERE id = 2")

conn.commit()

print("\nData Deleted")
```

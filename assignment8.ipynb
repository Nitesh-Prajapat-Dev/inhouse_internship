```python
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Nitesh", "Rahul", "Aman", "Rohit"]
})

# Second DataFrame
df2 = pd.DataFrame({
    "ID": [2, 3, 4, 5],
    "Marks": [85, 90, 78, 88]
})

print("DF1:\n")
print(df1)

print("\nDF2:\n")
print(df2)
```

```python
inner_merge = pd.merge(df1, df2, on="ID", how="inner")

print(inner_merge)
```

```python
left_join = pd.merge(df1, df2, on="ID", how="left")

print(left_join)
```

```python
right_join = pd.merge(df1, df2, on="ID", how="right")

print(right_join)
```

```python
df1_index = df1.set_index("ID")
df2_index = df2.set_index("ID")

joined_df = df1_index.join(df2_index)

print(joined_df)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Subject": ["Math", "Science", "English"],
    "Marks": [85, 90, 78]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Subject": ["Math", "Science", "English"],
    "Teacher": ["Sharma", "Verma", "Khan"]
})

merged = pd.merge(df1, df2, on=["ID", "Subject"])

print(merged)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "ID": [1, 2],
    "Name": ["Nitesh", "Rahul"]
})

df2 = pd.DataFrame({
    "ID": [3, 4],
    "Name": ["Aman", "Rohit"]
})

df3 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Marks": [85, 90, 78, 88]
})
```

```python
combined_df = pd.concat([df1, df2])

print(combined_df)
```

```python
final_df = pd.merge(combined_df, df3, on="ID")

print(final_df)
```

```python
df1 = df1.set_index("ID")
df3 = df3.set_index("ID")

result = df1.join(df3)

print(result)
```

```python
result = pd.merge(df1.reset_index(), df3.reset_index(), on="ID")

print(result)
```

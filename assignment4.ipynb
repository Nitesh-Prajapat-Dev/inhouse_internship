```python
def maximum(a, b, c):
    return max(a, b, c)

# Taking input
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

print("Maximum number is:", maximum(x, y, z))
```

```python
def distinct_list(lst):
    return list(set(lst))

# Example list
numbers = [1, 2, 2, 3, 4, 4, 5]

print("Original List:", numbers)
print("Distinct List:", distinct_list(numbers))
```

```python
def multiply_list(lst):
    result = 1

    for num in lst:
        result *= num

    return result

numbers = [2, 3, 4, 5]

print("List:", numbers)
print("Multiplication of all numbers:", multiply_list(numbers))
```

```python
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact *= i

    return fact

num = int(input("Enter a number: "))

if num < 0:
    print("Factorial does not exist for negative numbers")
else:
    print("Factorial =", factorial(num))
```

```python
def reverse_string(text):
    return text[::-1]

string = input("Enter a string: ")

print("Reversed String:", reverse_string(string))
```

```python
def check_range(num, start, end):
    if num in range(start, end + 1):
        print(num, "is within the range")
    else:
        print(num, "is outside the range")

number = int(input("Enter number: "))
start = int(input("Enter start range: "))
end = int(input("Enter end range: "))

check_range(number, start, end)
```

```python
def even_numbers(lst):
    print("Even numbers are:")

    for num in lst:
        if num % 2 == 0:
            print(num)

numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_numbers(numbers)
```

```python
def is_prime(num):

    if num <= 1:
        return False

    for i in range(2, num):
        if num % i == 0:
            return False

    return True

number = int(input("Enter a number: "))

if is_prime(number):
    print(number, "is a Prime Number")
else:
    print(number, "is not a Prime Number")
```

```python
def count_case(text):

    upper = 0
    lower = 0

    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Upper Case Letters:", upper)
    print("Lower Case Letters:", lower)

string = input("Enter a string: ")

count_case(string)
```

```python
import random
choices = ["stone", "paper", "scissor"]
computer = random.choice(choices)
player = input("Enter stone, paper or scissor: ").lower()
print("You chose:", player)
print("Computer chose:", computer)
if player == computer:
    print("Match Draw!")
elif player == "stone" and computer == "scissor":
    print("You Win!")
elif player == "paper" and computer == "stone":
    print("You Win!")
elif player == "scissor" and computer == "paper":
    print("You Win!")
elif player not in choices:
    print("Invalid Input!")
else:
    print("Computer Wins!")
```

```python
import requests

url = "https://api.exchangerate-api.com/v4/latest/USD"

response = requests.get(url)
data = response.json()

print("Base Currency:", data["base"])
print("INR Rate:", data["rates"]["INR"])
print("EUR Rate:", data["rates"]["EUR"])
print("GBP Rate:", data["rates"]["GBP"])
```

```python
import requests

# API URL
url = "https://ssd-api.jpl.nasa.gov/mdesign.api?lim=5&crit=1&year=2026&sb-group=neo"

# API Request
response = requests.get(url)

# JSON Data
data = response.json()

print("\nNASA JPL Small Body Mission Data\n")

print("API Version:", data["signature"]["version"])

print("\nAccessible Objects:\n")

# Print column names
print("Columns:", data["fields"])
print()

# Display proper data
for obj in data["data"]:

    print("Object Name:", obj[0])

    print("Launch Date:", obj[1])

    print("Arrival Date:", obj[3])

    print("Departure Velocity:", obj[6])

    print("Arrival Velocity:", obj[7])

    print("----------------------")
```
